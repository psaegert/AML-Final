```python
import numpy as np
import pandas as pd
from sklearn.svm import SVC

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix
```

# Data

```python
data = pd.read_csv('../data/data_geo_hosp.csv').drop(columns=['state_fips_code', 'county_fips_code', 'case_month'])
data
```

```python
Z = data.iloc[:, :].values.astype(float)
del data

X, y = Z[:, 2:], Z[:, :2]
del Z
print(f'{X.shape = }')
print(f'{y.shape = }')
```

# Preprocessing

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

# Training

```python
svc = SVC(kernel='linear', C=0.01)
svc.fit(X_train[:50_000], y_train[:50_000, 0].astype(bool))
```

# Evaluation

```python
y_pred = svc.predict(X_test)
```

```python
print(f'{accuracy_score(y_test[:, 0], y_pred) = }')
confusion_matrix(y_test[:, 0],y_pred)
```
