```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('../data/COVID-19_Case_Surveillance_Public_Use_Data.csv')
```

```python
data
```

```python
data = data.drop(columns=['cdc_case_earliest_dt ', 'cdc_report_dt', 'pos_spec_dt', 'onset_dt'])
data.head()
```

```python
data = data.dropna()
```

## Binary Categories

```python
# print(np.unique(data.current_status))
data.current_status = data.current_status.map({'Laboratory-confirmed case': 1, 'Probable Case': 0}).astype(int)
```

```python
# print(np.unique(data.sex))
data.sex = data.sex.map({'Female': 1, 'Male': 0, 'Missing': -1, 'Other': -1, 'Unknown': -1}).astype(int)
```

```python
# print(np.unique(data.hosp_yn))
data.hosp_yn = data.hosp_yn.map({'No': 0, 'Yes': 1, 'Missing': -1, 'Unknown': -1}).astype(int)
```

```python
# print(np.unique(data.icu_yn))
data.icu_yn = data.icu_yn.map({'No': 0, 'Yes': 1, 'Missing': -1, 'Unknown': -1}).astype(int)
```

```python
# print(np.unique(data.death_yn))
data.death_yn = data.death_yn.map({'No': 0, 'Yes': 1, 'Missing': -1, 'Unknown': -1}).astype(int)
```

```python
# print(np.unique(data.medcond_yn))
data.medcond_yn = data.medcond_yn.map({'No': 0, 'Yes': 1, 'Missing': -1, 'Unknown': -1}).astype(int)
```

## Age Groups

```python
data.age_group = data.age_group.map({'0 - 9 Years': 0, '10 - 19 Years': 10, '20 - 29 Years': 20, '30 - 39 Years': 30, '40 - 49 Years': 40, '50 - 59 Years': 50, '60 - 69 Years': 60, '70 - 79 Years': 70, '80+ Years': 80, 'Missing': 'missing'})

data = pd.concat([data, pd.get_dummies(data['age_group'], prefix='age')], axis=1)
data = data.drop(columns=['age_group'])
data.head()
```

## Races/Ethnicities

```python
data['race_missing'] = np.logical_or(data.race_ethnicity_combined.str.contains('Missing'), data.race_ethnicity_combined.str.contains('Unknown'))
data.head()
```

```python
data['race_native'] = data.race_ethnicity_combined.str.contains('American Indian/Alaska Native')
data['race_native_pacific'] = data.race_ethnicity_combined.str.contains('Native Hawaiian/Other Pacific Islander')
data['race_black'] = data.race_ethnicity_combined.str.contains('Black')
data['race_white'] = data.race_ethnicity_combined.str.contains('White')
data['race_asian'] = data.race_ethnicity_combined.str.contains('Asian')
data['race_hispanic'] = data.race_ethnicity_combined.str.contains('Hispanic/Latino')
data['race_other'] = data.race_ethnicity_combined.str.contains('Multiple/Other')

data = data.drop(columns=['race_ethnicity_combined'])
data
```

```python
# data.to_csv('../data/all_clean.csv')
```

```python
drop_mask = (
    (data.current_status == 0)
    # Features
    | (data.sex == -1) | (data.medcond_yn == -1) | (data.age_missing == 1) | (data.hosp_yn == -1) | (data.race_missing)
    # Dependent Variables
    | (data.hosp_yn == -1) | (data.death_yn == -1)
)
```

```python
data = data.loc[np.logical_not(drop_mask)]
data = data.reset_index()
data = data.astype(bool)
data = data.drop(columns=['index', 'age_missing', 'race_missing', 'icu_yn'])
data
```

```python
data.to_csv('../data/no_icu_clean.csv')
```
